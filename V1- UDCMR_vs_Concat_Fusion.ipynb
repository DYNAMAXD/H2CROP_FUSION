## Cell 3 вЂ” File Loader Utility

````python
def get_files_from_txt(txt_filename: str):
    """Read a .txt list and resolve each entry to its full .h5 path."""
    txt_path = DATA_LISTS_DIR / txt_filename
    if not txt_path.exists():
        print(f'{txt_path} not found вЂ” skipping.')
        return []

    matched, missing = [], 0
    with open(txt_path, 'r', encoding='utf-8') as f:
        for line in f:
            name = line.strip().lstrip('\ufeff')
            if not name:
                continue
            p = H5_DIR / f'{name}.h5'
            if p.exists():
                matched.append(p)
            else:
                missing += 1

    if missing:
        print(f'  {missing} entries in {txt_filename} had no matching .h5')
    return matched


train_files = get_files_from_txt('train_subset.txt')
val_files   = get_files_from_txt('val_subset.txt')
test_files  = get_files_from_txt('test_subset.txt')

print(f'\nLoaded  в†’  Train: {len(train_files)}  |  Val: {len(val_files)}  |  Test: {len(test_files)}')

if len(train_files) == 0:
    raise RuntimeError('No training files found. Check BASE_PATH and .txt contents.')
````

## Cell 4 вЂ” Dataset

````python
class DualStreamDataset(Dataset):
    # Per-channel Sentinel-2 statistics (10 bands Г— 12 time-steps = 120 channels after reshape)
    S2_MEAN = np.array([4259.26, 4215.91, 4103.17, 4499.33, 5117.05,
                        5324.11, 5487.41, 5417.51, 3758.27, 3095.50], dtype=np.float32)
    S2_STD  = np.array([3566.48, 3230.36, 3133.29, 3083.23, 2603.63,
                        2453.03, 2512.64, 2317.54, 1496.77, 1391.13], dtype=np.float32)

    def __init__(self, file_list, augment: bool = False):
        self.file_list = file_list
        self.augment   = augment

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        fpath = self.file_list[idx]
        with h5py.File(fpath, 'r') as f:
            s2    = f['S2_data'][:].astype(np.float32)
            label = f['label'][:]
            enmap = (f['EnMAP_data'][:].astype(np.float32)
                     if 'EnMAP_data' in f
                     else np.zeros((218, 64, 64), dtype=np.float32))

        # Normalise S2  (T, C, H, W) в†’ flatten time+bands в†’ (120, 192, 192)
        s2 = (s2 - self.S2_MEAN[None, :, None, None]) / (self.S2_STD[None, :, None, None] + 1e-8)
        s2_flat = s2.reshape(-1, 192, 192)          # (120, 192, 192)

        # Normalise EnMAP per-scene
        enmap = (enmap - enmap.mean()) / (enmap.std() + 1e-8)

        # Optional flip augmentation
        if self.augment and np.random.rand() > 0.5:
            s2_flat = np.flip(s2_flat, axis=-1).copy()
            enmap   = np.flip(enmap,   axis=-1).copy()

        # Majority label per hierarchy level  (ignore background=0)
        labs = []
        for li in range(4):
            vals = label[li][label[li] > 0]
            lab  = (int(np.bincount(vals).argmax()) - 1) if len(vals) > 0 else 0
            labs.append(max(0, lab))

        return (torch.from_numpy(s2_flat),
                torch.from_numpy(enmap),
                torch.tensor(labs, dtype=torch.long))


train_ds = DualStreamDataset(train_files, augment=True)
val_ds   = DualStreamDataset(val_files,   augment=False)
test_ds  = DualStreamDataset(test_files,  augment=False) if test_files else val_ds

# Balanced sampler on level-0 labels (scan first 200 samples for speed)
scan_n   = min(len(train_ds), 200)
l1_labels = [train_ds[i][2][0].item() for i in range(scan_n)]
cls_w    = 1.0 / (np.bincount(l1_labels, minlength=HIERARCHY[0]) + 1e-6)
s_weights = cls_w[l1_labels]
sampler  = WeightedRandomSampler(s_weights, num_samples=len(s_weights), replacement=True)

train_dl = DataLoader(train_ds, batch_size=BATCH_SIZE, sampler=sampler,  num_workers=0)
val_dl   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False,    num_workers=0)
test_dl  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False,    num_workers=0)

print(f'Batches  в†’  Train: {len(train_dl)}  |  Val: {len(val_dl)}  |  Test: {len(test_dl)}')
````

##  Cell 5 вЂ” Shared Building Blocks

````python
class TransformerBlock(nn.Module):
    """Pre-LN Transformer block with 4-head MHA and feed-forward."""
    def __init__(self, d_model: int, n_heads: int = 4, ffn_mult: int = 4, drop: float = 0.1):
        super().__init__()
        self.ln1  = nn.LayerNorm(d_model)
        self.attn = nn.MultiheadAttention(d_model, n_heads, dropout=drop, batch_first=True)
        self.ln2  = nn.LayerNorm(d_model)
        self.mlp  = nn.Sequential(
            nn.Linear(d_model, d_model * ffn_mult),
            nn.GELU(),
            nn.Dropout(drop),
            nn.Linear(d_model * ffn_mult, d_model),
        )
        self.drop = nn.Dropout(drop)

    def forward(self, x):
        x = x + self.drop(self.attn(self.ln1(x), self.ln1(x), self.ln1(x))[0])
        return x + self.drop(self.mlp(self.ln2(x)))
````

## Cell 6 вЂ” Baseline: Simple Concat-Fusion Model

````python
class ConcatFusionNet(nn.Module):
    """
    Baseline dual-stream model.
    Fusion = torch.cat([s2_pool, hsi_pool], dim=-1)  в†’ shared MLP  в†’ cascade heads.
    No gating, no TDA alignment.
    """
    def __init__(self, hierarchy: dict, d_model: int = 256):
        super().__init__()
        # S2 encoder
        self.s2_patch  = nn.Conv2d(120, d_model, kernel_size=16, stride=16)
        self.s2_blocks = nn.Sequential(*[TransformerBlock(d_model) for _ in range(2)])

        # HSI encoder
        self.hsi_compress = nn.Sequential(
            nn.Conv2d(1, 16, 3, 2, 1), nn.GELU(),
            nn.AdaptiveAvgPool2d(4), nn.Flatten(),
            nn.Linear(16 * 16, d_model)
        )
        self.hsi_blocks = nn.Sequential(*[TransformerBlock(d_model) for _ in range(2)])

        # Simple concat в†’ projection
        self.fusion_proj = nn.Sequential(
            nn.Linear(d_model * 2, d_model),
            nn.GELU(),
            nn.Dropout(0.3),
        )

        # Cascade classification heads
        self.heads = nn.ModuleList()
        prev = 0
        for n_cls in hierarchy.values():
            self.heads.append(nn.Linear(d_model + prev, n_cls))
            prev = n_cls

    def forward(self, s2, hsi):
        # S2 stream
        s2_m = self.s2_patch(s2)                              # (B, d, 12, 12)
        s2_t = self.s2_blocks(s2_m.flatten(2).transpose(1, 2))  # (B, 144, d)
        s2_p = s2_t.mean(1)                                   # (B, d)

        # HSI stream
        B, C, H, W = hsi.shape
        hsi_t = self.hsi_blocks(
            self.hsi_compress(hsi.view(B * C, 1, H, W)).view(B, C, -1)
        )                                                      # (B, C, d)
        hsi_p = hsi_t.mean(1)                                 # (B, d)

        # Naive concat fusion
        fused  = self.fusion_proj(torch.cat([s2_p, hsi_p], dim=-1))

        # Cascade heads
        logits, prev_s = [], None
        for head in self.heads:
            inp = fused if prev_s is None else torch.cat([fused, prev_s], dim=-1)
            l   = head(inp)
            logits.append(l)
            prev_s = torch.softmax(l.detach(), dim=-1)
        return logits, None   # None в†’ no gate weights


print('ConcatFusionNet defined')
````

## Cell 7 вЂ” Proposed: UDCMR-Net (TDA + Bayesian Gate)

````python
class UDCMR_Net(nn.Module):
    """
    UDCMR-Net:
      вЂў TDA  вЂ” Transformer-guided Deformable Alignment (offset prediction)
      вЂў CMRG вЂ” Cross-Modal Routing Gate  (Bayesian soft gating over 3 routes)
      вЂў Cascade heads propagate soft predictions down the label hierarchy
    """
    def __init__(self, hierarchy: dict, d_model: int = 256):
        super().__init__()
        # S2 encoder
        self.s2_patch  = nn.Conv2d(120, d_model, kernel_size=16, stride=16)
        self.s2_blocks = nn.Sequential(*[TransformerBlock(d_model) for _ in range(2)])

        # HSI encoder
        self.hsi_compress = nn.Sequential(
            nn.Conv2d(1, 16, 3, 2, 1), nn.GELU(),
            nn.AdaptiveAvgPool2d(4), nn.Flatten(),
            nn.Linear(16 * 16, d_model)
        )
        self.hsi_blocks = nn.Sequential(*[TransformerBlock(d_model) for _ in range(2)])

        # TDA вЂ” deformable alignment (predict 2-D offsets from concat feature maps)
        self.tda_offsets = nn.Sequential(
            nn.Conv2d(d_model * 2, d_model, 3, 1, 1),
            nn.GroupNorm(8, d_model),
            nn.GELU(),
            nn.Conv2d(d_model, 2, 3, 1, 1),      # в†’ (B, 2, 12, 12)  (dx, dy)
        )

        # Bayesian routing gate  в†’ 3 mixing weights  [S2, HSI, joint]
        self.gate = nn.Sequential(
            nn.Linear(d_model * 2, d_model // 2),
            nn.GELU(),
            nn.Dropout(0.3),
            nn.Linear(d_model // 2, 3),
        )

        # Cascade classification heads
        self.heads = nn.ModuleList()
        prev = 0
        for n_cls in hierarchy.values():
            self.heads.append(nn.Linear(d_model + prev, n_cls))
            prev = n_cls

    def forward(self, s2, hsi):
        # в”Ђв”Ђ S2 stream в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
        s2_m = self.s2_patch(s2)                              # (B, d, 12, 12)
        s2_t = self.s2_blocks(s2_m.flatten(2).transpose(1, 2))  # (B, 144, d)
        s2_p = s2_t.mean(1)                                   # (B, d)

        # в”Ђв”Ђ HSI stream в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
        B, C, H, W = hsi.shape
        hsi_t = self.hsi_blocks(
            self.hsi_compress(hsi.view(B * C, 1, H, W)).view(B, C, -1)
        )                                                      # (B, C, d)
        hsi_cls = hsi_t.mean(1)                               # (B, d)

        # в”Ђв”Ђ TDA alignment в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
        # Reshape first 144 HSI tokens to spatial (12Г—12) and align with S2 map
        hsi_spatial = F.interpolate(
            hsi_t[:, :144].transpose(1, 2).view(B, -1, 12, 12),
            size=(12, 12)
        )                                                      # (B, d, 12, 12)
        offsets = self.tda_offsets(
            torch.cat([s2_m, hsi_spatial], dim=1)
        )                                                      # (B, 2, 12, 12)

        # в”Ђв”Ђ Bayesian gate в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
        gate_in = torch.cat([s2_p, hsi_cls], dim=-1)          # (B, 2d)
        w = F.softmax(self.gate(gate_in), dim=-1)              # (B, 3)

        # в”Ђв”Ђ Weighted fusion  [S2-only | HSI-only | joint-CLS] в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
        joint = s2_t.mean(1)   # could be cross-attn; here: average as routing proxy
        fused = (w[:, 0:1] * s2_p
               + w[:, 1:2] * hsi_cls
               + w[:, 2:3] * joint)                           # (B, d)

        # в”Ђв”Ђ Cascade heads в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
        logits, prev_s = [], None
        for head in self.heads:
            inp = fused if prev_s is None else torch.cat([fused, prev_s], dim=-1)
            l   = head(inp)
            logits.append(l)
            prev_s = torch.softmax(l.detach(), dim=-1)

        return logits, w   # w в†’ gate weights for analysis


print('UDCMR_Net defined')
````

## Cell 8 вЂ” Training & Evaluation Utilities

````python
def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss, n_batches = 0.0, 0
    for s2, hsi, lbl in loader:
        s2, hsi, lbl = s2.to(device), hsi.to(device), lbl.to(device)
        optimizer.zero_grad()
        out, _ = model(s2, hsi)
        loss = sum(criterion(out[i], lbl[:, i]) for i in range(4))
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item()
        n_batches  += 1
    return total_loss / max(n_batches, 1)


@torch.no_grad()
def evaluate(model, loader, criterion):
    model.eval()
    total_loss  = 0.0
    correct     = [0] * 4
    total       = 0
    gate_w_list = []   # collect gate weights if available

    for s2, hsi, lbl in loader:
        s2, hsi, lbl = s2.to(device), hsi.to(device), lbl.to(device)
        out, gate_w  = model(s2, hsi)
        loss = sum(criterion(out[i], lbl[:, i]) for i in range(4))
        total_loss += loss.item()

        for i in range(4):
            preds = out[i].argmax(dim=-1)
            correct[i] += (preds == lbl[:, i]).sum().item()
        total += lbl.size(0)

        if gate_w is not None:
            gate_w_list.append(gate_w.cpu())

    acc_per_level = [c / max(total, 1) for c in correct]
    avg_loss      = total_loss / max(len(loader), 1)
    gate_mean     = torch.cat(gate_w_list).mean(0).numpy() if gate_w_list else None
    return avg_loss, acc_per_level, gate_mean


print('Training utilities ready')
````

## Cell 9 вЂ” Run Training for Both Models

````python
criterion = nn.CrossEntropyLoss()

def run_experiment(ModelClass, name: str, color: str):
    print(f'\n{'='*60}')
    print(f'  Training  в†’  {name}')
    print(f'{'='*60}')

    model = ModelClass(HIERARCHY, d_model=D_MODEL).to(device)
    n_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f'  Parameters: {n_params:,}')

    optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS)

    history = defaultdict(list)
    best_val_acc = 0.0
    best_ckpt    = SAVE_DIR / f'{name.replace(" ", "_")}_best.pt'
    gate_history = []

    for epoch in range(1, NUM_EPOCHS + 1):
        t0 = time.time()

        tr_loss                       = train_one_epoch(model, train_dl, optimizer, criterion)
        va_loss, va_acc, gate_w_mean  = evaluate(model, val_dl, criterion)
        scheduler.step()

        history['train_loss'].append(tr_loss)
        history['val_loss'].append(va_loss)
        for li in range(4):
            history[f'val_acc_L{li}'].append(va_acc[li])

        if gate_w_mean is not None:
            gate_history.append(gate_w_mean)

        # Save best checkpoint (based on level-0 accuracy)
        if va_acc[0] > best_val_acc:
            best_val_acc = va_acc[0]
            torch.save({
                'epoch'     : epoch,
                'model_state': model.state_dict(),
                'opt_state' : optimizer.state_dict(),
                'val_acc_L0': best_val_acc,
                'history'   : dict(history),
            }, best_ckpt)

        elapsed = time.time() - t0
        acc_str = '  '.join(f'L{i}:{va_acc[i]*100:.1f}%' for i in range(4))
        print(f'  Ep {epoch:02d}/{NUM_EPOCHS}  tr_loss={tr_loss:.4f}  va_loss={va_loss:.4f}  '
              f'{acc_str}  ({elapsed:.1f}s)')

    # Final checkpoint
    final_ckpt = SAVE_DIR / f'{name.replace(" ", "_")}_final_ep{NUM_EPOCHS}.pt'
    torch.save(model.state_dict(), final_ckpt)
    print(f'\n  Best   model saved в†’ {best_ckpt.name}')
    print(f'  Final  model saved в†’ {final_ckpt.name}')

    return dict(history), gate_history, model


# в”Ђв”Ђ Run both experiments в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
concat_hist, _,            concat_model  = run_experiment(ConcatFusionNet, 'Concat_Fusion', '#3B82F6')
udcmr_hist,  udcmr_gates,  udcmr_model   = run_experiment(UDCMR_Net,       'UDCMR_Net',     '#10B981')
````

## Cell 11 вЂ” Confusion Matrix on Test Set (Level 0)

````python
@torch.no_grad()
def get_preds_labels(model, loader, level=0):
    model.eval()
    all_preds, all_labels = [], []
    for s2, hsi, lbl in loader:
        s2, hsi = s2.to(device), hsi.to(device)
        out, _  = model(s2, hsi)
        all_preds.append(out[level].argmax(-1).cpu())
        all_labels.append(lbl[:, level])
    return torch.cat(all_preds).numpy(), torch.cat(all_labels).numpy()


c_preds, c_true = get_preds_labels(concat_model, test_dl, level=0)
u_preds, u_true = get_preds_labels(udcmr_model,  test_dl, level=0)

fig, axes = plt.subplots(1, 2, figsize=(16, 6), facecolor=C_BG)

for ax, preds, true, title, color in [
    (axes[0], c_preds, c_true, 'Concat Fusion вЂ” Level-0 CM', C_CONCAT),
    (axes[1], u_preds, u_true, 'UDCMR-Net вЂ” Level-0 CM',     C_UDCMR),
]:
    cm   = confusion_matrix(true, preds)
    cm_n = cm.astype(float) / (cm.sum(axis=1, keepdims=True) + 1e-6)   # row-normalise
    sns.heatmap(cm_n, ax=ax, cmap='Blues', annot=True, fmt='.2f',
                linewidths=0.5, linecolor=C_GRID,
                cbar_kws={'shrink': 0.7})
    ax.set_title(title, fontsize=13, fontweight='bold', color=C_TEXT)
    ax.set_xlabel('Predicted', color=C_TEXT)
    ax.set_ylabel('True',      color=C_TEXT)
    ax.tick_params(colors=C_TEXT)
    ax.set_facecolor(C_PANEL)

plt.tight_layout()
plt.savefig(str(SAVE_DIR / 'confusion_matrices.png'),
            dpi=150, bbox_inches='tight', facecolor=C_BG)
plt.show()
print(f'Saved в†’ {SAVE_DIR / "confusion_matrices.png"}')
````
